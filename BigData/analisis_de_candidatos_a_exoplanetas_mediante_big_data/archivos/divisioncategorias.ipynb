```pyspark
from pyspark.sql import SparkSession
# 1. Crear sesión Spark
spark = SparkSession.builder.appName("DividirExoplanetasPorClasificación").getOrCreate()
```

```pyspark
# 2. Leer el dataset ignorando comentarios (líneas que comienzan por #)
df = spark.read.option("header", True)\
               .option("inferSchema", True)\
               .option("comment", "#")\
               .csv(input_path)
```

```pyspark
# 3. Filtrar por cada tipo de clasificación
df_confirmed = df.filter(df["koi_disposition"] == "CONFIRMED")
df_candidates = df.filter(df["koi_disposition"] == "CANDIDATE")
df_false_positives = df.filter(df["koi_disposition"] == "FALSE POSITIVE")
```

```pyspark
# 4. Guardar cada dataframe como CSV en su carpeta correspondiente en S3
df_confirmed.write.mode("overwrite").option("header", True).csv("s3://origendatoskoi/confirmed/confirmedkoi.csv")
df_candidates.write.mode("overwrite").option("header", True).csv("s3://origendatoskoi/candidates/candidateskoi.csv")
df_false_positives.write.mode("overwrite").option("header", True).csv("s3://origendatoskoi/falsepositive/falsepositivekoi.csv")
```
